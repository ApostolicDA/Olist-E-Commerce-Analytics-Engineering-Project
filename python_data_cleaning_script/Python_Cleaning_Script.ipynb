```python
df['seller_city']=df['seller_city'].replace ({
    'são paulo': 'sao paulo',
    'sao pauo': 'sao paulo',
    'sao paulop': 'sao paulo',
    'sao paulo / sao paulo': 'sao paulo',
    'sao paulo - sp': 'sao paulo',
    'sao  paulo': 'sao paulo',
    'mogi das cruses': 'mogi das cruzes',
    'ferraz de  vasconcelos': 'ferraz de vasconcelos',
    'cascavael': 'cascavel',
    'ribeirao pretp': 'ribeirao preto',
    'rio de janeiro \\rio de janeiro': 'rio de janeiro',
    'rio de janeiro / rio de janeiro': 'rio de janeiro',
    'rio de janeiro, rio de janeiro, brasil': 'rio de janeiro',
    'sbc/sp': 'sao bernardo do campo',
    'sp / sp': 'sao paulo',
    'sao pedro dos pinhas': 'sao jose dos pinhais',
    'balenario camboriu': 'balneario camboriu',
    'são paulo': 'sao paulo',
    'sao bernardo do capo': 'sao bernardo do campo'
})
```

```python
df['seller_city'].unique()
```

```python
df['seller_city'] = (
    df['seller_city']
    .str.lower()
    .str.strip()
)
```

```python
#  Remove state info after slash, dash, comma, parentheses
df['seller_city'] = (
    df['seller_city']
    .str.split(r'[/,-]|\\(|\\)', expand=False)
    .str[0]
    .str.strip()
)
```

```python
#  Remove double spaces
df['seller_city'] = df['seller_city'].str.replace(r'\s+', ' ', regex=True)
```

```python
#  Remove obvious non-city values
df = df[~df['seller_city'].str.contains(r'@|\d{5,}', regex=True)]
```

```python
# 5 Fix common typos
replacements = {
    'garulhos': 'guarulhos',
    'robeirao preto': 'ribeirao preto',
    'riberao preto': 'ribeirao preto',
    'sao paluo': 'sao paulo',
    'belo horizont': 'belo horizonte',
    'ao bernardo do campo': 'sao bernardo do campo',
    'scao jose do rio pardo': 'sao jose do rio pardo',
    's jose do rio preto': 'sao jose do rio preto',
    'sao jose do rio pret': 'sao jose do rio preto',
    'sao jose dos pinhas': 'sao jose dos pinhais',
}

df['seller_city'] = df['seller_city'].replace(replacements)
```

```python
df['seller_city'].unique()
```

```python
# Remove state-only entries
states = ['bahia', 'tocantins', 'santa catarina', 'parana', 'minas gerais', 'sp']
df = df[~df['seller_city'].isin(states)].copy()

# Remove state suffixes
df['seller_city'] = (
    df['seller_city']
    .str.replace(r'\s+(sp|rj|df)$', '', regex=True)
)

# Fix remaining typos
final_replacements = {
    'floranopolis': 'florianopolis',
    'sando andre': 'santo andre',
    'portoferreira': 'porto ferreira',
    'sbc': 'sao bernardo do campo',
    'centro': None,
    "santa barbara d´oeste": "santa barbara d'oeste"
}

df['seller_city'] = df['seller_city'].replace(final_replacements)

# Drop None values
df = df.dropna(subset=['seller_city'])
```

```python
df['seller_city'].unique()
```

```python
df['seller_city'] = df['seller_city'].str.replace(" d oeste", " d'oeste")
```

```python
df['seller_city'] = df['seller_city'].replace(
    {"arraial d'ajuda (porto seguro)": "porto seguro"}
)
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.to_csv(r"C:\Users\gadis\Downloads\olist\Clean Datasets\sellers.csv",index=False, lineterminator="\n")
```
